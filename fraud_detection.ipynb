Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
#loading the dataset to a pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
#last 5 rows of the datset
credit_card_data.tail()
```

```python
#datset informations
credit_card_data.info()
```

```python
#checking number of missing values in each column
#in presence of any missing values you have to assign them numbers, in this datset we don't have any missing values
credit_card_data.isnull().sum()
```

```python
#checking distribution of the dataset
 #(0: legit transaction, 1: fraud)
credit_card_data['Class'].value_counts()
```

```python
#since our data is very unbalanced, we have to process it

#separating data for analysis

legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
#ststistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both classes
credit_card_data.groupby('Class').mean()
```

UNDER SAMPLING

we will build a sample dataset containing similar distribution of normal and fraud transactions


```python
legit_sample = legit.sample(n=492)
```

Concatenate two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis =0)
#value should be added rowwise so we give axis = 0, for adding in column we would have given axis = 1
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

the mean values are not very different, so we can tell that the sample is kinda working the same

SPLITTING the datset into features and targets

```python
X= new_dataset.drop(columns='Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2, stratify=Y, random_state = 2)
 
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training

(generally we us LR for binary classification)

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the LR model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
print("Training data accuracy is : ", training_data_accuracy)
```

```python
 #accuracy on the test data
X_test_predict = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predict, Y_test)
print("Testing data accuracy is : ", test_data_accuracy) 
```
